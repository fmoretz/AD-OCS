```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
adm1_input = pd.read_csv(r'C:\Users\fede1\OneDrive - Politecnico di Milano\Documenti\GitHub\AD-OCS\AD-OCS\PyADM1\PyADM1_dyn\thoni_digester_influent_dyn.csv', index_col=0)
adm1_input['XT'] = adm1_input['X_xc']+adm1_input['X_ch']+adm1_input['X_pr']+adm1_input['X_li']
adm1_input['S1'] = adm1_input['S_su'] + adm1_input['S_aa'] + adm1_input['S_fa'] 
adm1_input['S2'] = (adm1_input['S_va']/208 + adm1_input['S_bu']/160 + adm1_input['S_pro']/112 + adm1_input['S_ac']/64)*1000
adm1_input['C'] = adm1_input['S_IC']*1000
adm1_input['N'] = adm1_input['S_IN']*1000

# Convert ADM1 variable to AD_OCS variables
ad_ocs_input = pd.DataFrame()
ad_ocs_input['Qin'] = adm1_input['Q']
ad_ocs_input['S1in'] = adm1_input['S1']
ad_ocs_input['S2in'] = adm1_input['S2']
ad_ocs_input['XTin'] = adm1_input['XT']
ad_ocs_input['XTin'] = adm1_input['XT']
ad_ocs_input['Cin'] = adm1_input['C']
ad_ocs_input['Nin'] = adm1_input['N']

# Create the new dataframe
A = ad_ocs_input.index.to_numpy
# round to 2 decimal places
ad_ocs_input = ad_ocs_input.round(2)
ad_ocs_input['time']=ad_ocs_input.index
ad_ocs_input['time']=np.floor(ad_ocs_input['time']).astype(int)+1
ad_ocs_input.drop_duplicates(subset='time', keep='first', inplace=True)
ad_ocs_input.set_index('time', inplace=True)
ad_ocs_input
```

```python
ad_ocs_input['Qin_dev'] = ad_ocs_input['Qin']/ad_ocs_input['Qin'].iloc[1]
ad_ocs_input['S1in_dev'] = ad_ocs_input['S1in']/ad_ocs_input['S1in'].iloc[0]
ad_ocs_input['S2in_dev'] = ad_ocs_input['S2in']/ad_ocs_input['S2in'].iloc[0]
ad_ocs_input['XTin_dev'] = ad_ocs_input['XTin']/ad_ocs_input['XTin'].iloc[0]
ad_ocs_input['Cin_dev'] = ad_ocs_input['Cin']/ad_ocs_input['Cin'].iloc[0]
ad_ocs_input['Nin_dev'] = ad_ocs_input['Nin']/ad_ocs_input['Nin'].iloc[0]
ad_ocs_input.round(5).fillna(0)

deviations = pd.DataFrame()
deviations['Qin'] = ad_ocs_input['Qin_dev'][1:]
deviations['S1in'] = ad_ocs_input['S1in_dev'][1:]
deviations['S2in'] = ad_ocs_input['S2in_dev'][1:]
deviations['XTin'] = ad_ocs_input['XTin_dev'][1:]
deviations['Cin'] = ad_ocs_input['Cin_dev'][1:]
deviations['Nin'] = ad_ocs_input['Nin_dev'][1:]
deviations['time'] = ad_ocs_input.index[1:]
deviations.fillna(0, inplace=True)
#deviations.to_excel(r"C:\Users\fede1\OneDrive - Politecnico di Milano\Documenti\GitHub\AD-OCS\Working_data\thoni_deviations.xlsx", index=False, header=True, sheet_name='Deviations')
```
